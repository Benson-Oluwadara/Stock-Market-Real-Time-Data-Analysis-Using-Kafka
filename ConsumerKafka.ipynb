```python
import pandas as pd
from kafka import KafkaConsumer
import json
from time import sleep
from json import dumps,loads
from s3fs import S3FileSystem
consumer=KafkaConsumer('demotest',bootstrap_servers=['44.203.154.25:9092'],
                       value_deserializer=lambda x:loads(x.decode('utf-8')))


#for c in consumer:
#    print(c.value)

s3=S3FileSystem()

for count,i in enumerate(consumer):
    with s3.open("s3://kafkastockmarketdara/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value,file)
        #print(count,i.value)
```
